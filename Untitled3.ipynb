```python
import json as j
import pandas as pd
import re
import numpy as np
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, chi2
from nltk.stem import WordNetLemmatizer
```

```python
new = pd.read_csv('test.csv')
new
```

```python
remove = ['Classes']
new = new[new.columns.difference(remove)]
new =  new.apply(lambda x: x.astype(str).str.lower())
new
```

```python
corpus =[]
for i in range(0,1414):
    review = re.sub('[^a-zA-Z]',' ',new['question_description'][i])
    review = review +' '+ new['Tag1'][i]+' '+new['Tag2'][i]
    review = review.split()
    porter = PorterStemmer()
    #review = [porter.stem(t) for t in review]
    ps = WordNetLemmatizer()
    review = [ps.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review =' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
y = pd.read_csv("category.csv")
y = y[0:1414]
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size=0.25)
```

```python
pipeline = Pipeline([('vect', TfidfVectorizer(ngram_range=(0, 1), stop_words="english", sublinear_tf=True)),
                     ('chi',  SelectKBest(chi2, k=800)),
                     ('clf', SVC(kernel ='linear'))])


model = pipeline.fit(X_train, y_train)
```

```python
vectorizer = model.named_steps['vect']
chi = model.named_steps['chi']
clf = model.named_steps['clf']
```

```python
feature_names = vectorizer.get_feature_names()
feature_names = [feature_names[i] for i in chi.get_support(indices=True)]
feature_names = np.asarray(feature_names)
```

```python
feature_names
```

```python
print("accuracy score: " + str(model.score(X_test, y_test)))
```

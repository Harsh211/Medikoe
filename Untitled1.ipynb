```python
import pandas as pd
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest,chi2
```

```python
data = pd.read_csv('medikoe.csv')
data
```

```python
data.shape
```

```python
remove = ['created_at','creator_id','id','private','question_title','question_url','status','updated_at','view_count','category','topic']
data = data[data.columns.difference(remove)]
data =  data.apply(lambda x: x.astype(str).str.lower())
data
```

```python
corpus =[]
for i in range(0,1800):
    review = re.sub('[^a-zA-Z]',' ',data['question_description'][i])
    review = review.split()
    porter = PorterStemmer()
    #review = [porter.stem(t) for t in review]
    ps = WordNetLemmatizer()
    review = [ps.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review =' '.join(review)
    corpus.append(review)
    print(review)
```

```python
y = pd.read_csv("question.csv")
```

```python
X_train, X_test, y_train, y_test = train_test_split(corpus,y[0:1800],test_size=.25)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
pipeline = Pipeline([('vect', TfidfVectorizer(ngram_range=(0, 5), stop_words="english", sublinear_tf=True, max_features=1700)),
                     ('chi',  SelectKBest(chi2, k=1500)),
                     ('clf', SVC(kernel ='linear',random_state=0))])


model = pipeline.fit(X_train, y_train)
```

```python
print("accuracy score: " + str(model.score(X_test, y_test)))
```

```python
print(model.predict(['marijuana effect']))
```

```python
result =[]
for i in range(0,1800):
    test = data['question_description'][i]
    test = test.split()
    porter = PorterStemmer()
    #test = [porter.stem(t) for t in test]
    lemma = WordNetLemmatizer()
    test = [lemma.lemmatize(word) for word in test if not word in set(stopwords.words('english'))]
    test =' '.join(test)
    result.append(test)
```

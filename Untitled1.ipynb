```python
import pandas as pd
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest,chi2
```

```python
data = pd.read_json('resp', orient='coloumns')
data = pd.read_json((data['questions']).to_json(), orient='index')
data
```

```python
data.shape
```

```python
remove = ['created_at','creator_id','id','private','question_title','question_url','status','updated_at','view_count','category','topic']
data = data[data.columns.difference(remove)]
data =  data.apply(lambda x: x.astype(str).str.lower())
data
```

```python
corpus =[]
for i in range(0,1200):
    review = re.sub('[^a-zA-Z]',' ',data['question_description'][i])
    review = review.split()
    ps = WordNetLemmatizer()
    review = [ps.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review =' '.join(review)
    corpus.append(review)
    print(review)
```

```python
y = pd.read_csv("category.csv")
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(corpus,y,test_size=.2)
```

```python
pipeline = Pipeline([('vect', CountVectorizer(max_features=800)),
                     ('chi',  SelectKBest(chi2, k=300)),
                     ('clf', SVC(kernel ='linear',random_state=0))])


model = pipeline.fit(X_train, y_train)
```

```python
print("accuracy score: " + str(model.score(X_test, y_test)))
```

```python
print(model.predict(['Skin burning']))
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_pred,y_test)

from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_pred)
```

```python
plt.close()
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Wistia)
classNames = ['Negative','Positive']
plt.title('SVM Linear Kernel Confusion Matrix - Test Data')
plt.ylabel('True label')
plt.xlabel('Predicted label')
tick_marks = np.arange(len(classNames))
plt.xticks(tick_marks, classNames, rotation=45)
plt.yticks(tick_marks, classNames)
s = [['TN  ','FP  '], ['FN  ', 'TP  ']]
 
for i in range(2):
    for j in range(2):
        plt.text(j,i, str(s[i][j])+" = "+str(cm[i][j]))
plt.show()
```
